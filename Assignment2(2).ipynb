b) Second part of the second question wants us to implement how the agent interacts with the environment (i.e., the step function, restart function, init function etc.)

We will explain the three functions above before making the code to implement them. 
The step() funcitonreturns the current observations and the one-step reward
The reset() function sets the environment to an initial state and computes the initial value of the observation.
The init() is executed when the class is being initialized. 

```python
import gym
from gym import spaces

class CustomEnv(gym.Env):
    #We will begin with init function. We include the number of balls, bins (start position) and number of actions (-1, 0, 1)
    #symbolizing whether place ball right, same or left.
    def __init__(self):
        self.bins = [5] * 10
        self.balls = list(range(10))
        self.actions = [-1, 0, 1]
        self.observation_space = spaces.Box(low=0, high=1, shape=(10, 10), dtype=np.int32)
        self.action_space = spaces.Discrete(len(self.actions))
            
    def reset(self):
        self.bins = [5] * 10
        self.balls = list(range(10))
        return self._get_state()
    
    def step(self, action):
        ball = self.balls.pop(0)
        binpos = ball
        if action == 0:
            pass
        elif action == 1 and binpos < 9:
            binpos += 1
        elif action == 2 and binpos > 0:
            binpos -= 1
        self.bins[ball] = binpos
        reward = self._get_reward()
        done = len(self.balls) == 0
        return self._get_state(), reward, done, {}
    
    def _get_state(self):
        state = np.zeros((10, 10), dtype=np.int32)
        for i, ball in enumerate(self.balls):
            state[ball, self.bins[ball]] = 1
        return state
    
    def _get_reward(self):
        return -np.sum(np.abs(np.array(self.bins) - np.mean(self.bins)))
    
#We continue on to train the RL agent.
env = CustomEnv()
C = np.zeros((100000000, env.action_space.n))
a = 0.001
g = 0.9
for i in range(10):
    state = env.reset()
    action = np.argmax(C[i])
    if random.random() < 0.001:
        action = random.choice(range(env.action_space.n))
    done = False
    while not done:
        new_state, reward, done, _ = env.step(action)
        new_action = np.argmax(C[i])
        if random.random() < 0.1:
            new_action = random.choice(range(env.action_space.n))
        C[i+1, action] += a * (reward + g * np.max(C[i, :]) - C[i, action])

        state = new_state
        action = new_action

print(env.bins)
```

```python
def check_env(env, warn=True):
    if not isinstance(env, gym.Env):
        if warn:
            print("Warning: env is not an instance of gym.Env")
        return False

    if not isinstance(env.observation_space, spaces.Box):
        if warn:
            print("Warning: observation space is not an instance of spaces.Box")
        return False

    if not isinstance(env.action_space, spaces.Discrete):
        if warn:
            print("Warning: action space is not an instance of spaces.Discrete")
        return False

    return True
```
